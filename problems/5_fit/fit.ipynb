```python
from numpy import *from scipy.optimize import curve_fitfrom matplotlib.pyplot import *
```

Let us try to fit a curve for a given data set. Let us go ahead and import the data set. The data is stored in text files 'x.txt' and 'y.txt'. 'loadtxt' is a numpy function which gathers data from a text file and stores it in a variable.

```python
x, y = loadtxt('x.txt'), loadtxt('y.txt')
```

It is always a good idea to plot and visualize the data before trying to guess a model.

```python
plot(x,y,'o')show()
```

It looks somewhat like an exponential function. Our best guess was a function like the one below. We need to define a python funcion which acts as our model.

```python
def model(x, c, m):	return c*(1 + exp(-m*x))
```

The above goes to show that the curve 'might' fit the data by simply changing the parameters c and m. It may not fit well in most applications and you might have to guess a couple of times before arriving at the best fit function. Let us proceed to fit the data.

```python
params, covariance  = curve_fit(model, x, y)
```

'params' contains those values of c and m which seem to fit the data the best. 'covariance' is a matrix which gives us an idea of the dependency between the two variables. We won't be using covariance in this example.

```python
c, m = params[0], params[1]
```

Now let us see how well this fits our data. Plots are your friends, use them to your advantage.

```python
plot(x,y,'-')plot(x,model(x, c, m))show()
```
