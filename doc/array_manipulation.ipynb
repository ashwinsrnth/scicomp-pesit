### Loading an image onto Python:

```python
i = loadtxt('pup.txt')
```

```python
imshow(i)
```

```python
print i
```

This is an array of shape [300, 332, 4]. This means, primarily, that our image is 300 x 322 pixels. Visualize this array as 4 layers of matrices sized 300 x 322. Each matrix represents, respectively:1. The intensity of red at the corresponding position2. The intensity of blue at the corresponding position3. The intensity of green at the corresponding position4. The 'transparency' (alpha) at each position.Since this image is grayscale, only one of the first three layers will suffice to describe the imagecompletely. So, start with:`i = i[:,:,0]`and set the colormap to gray:`gray()`Now, the matrix `i` describes the image, with each element of the matrix corresponding to the intensityof gray (0 = black, 255 = white) at that particular position (pixel).You are ready to solve the problem

```python
i = i[:,:,0]
```
